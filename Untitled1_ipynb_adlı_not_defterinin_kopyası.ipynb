```python
from transformers import pipeline

# Load pre-trained sentiment analysis model
classifier = pipeline("sentiment-analysis")

def analyze_sentiment(text):
    result = classifier(text)
    return result[0]

# Test example
text = "Stock market surges after strong earnings report"
prediction = analyze_sentiment(text)

print("Text:", text)
print("Sentiment:", prediction["label"])
print("Confidence:", round(prediction["score"], 3))
```

```python
user_input = input("Enter a financial news headline: ")
prediction = analyze_sentiment(user_input)

print("\nSentiment:", prediction["label"])
print("Confidence:", round(prediction["score"], 3))
```

```python
from transformers import pipeline

# Load finance-specific sentiment model
classifier = pipeline("sentiment-analysis", model="ProsusAI/finbert")

def analyze_sentiment(text):
    result = classifier(text)
    return result[0]

# Test example
text = "Company shares fall after disappointing quarterly earnings"
prediction = analyze_sentiment(text)

print("Text:", text)
print("Sentiment:", prediction["label"])
print("Confidence:", round(prediction["score"], 3))
```

```python
import pandas as pd

def market_impact_score(prediction):
    label = prediction["label"]
    score = prediction["score"]

    if label == "positive":
        return score
    elif label == "negative":
        return -score
    else:
        return 0

# Example test
text = "Federal Reserve signals potential rate cuts next quarter"
prediction = analyze_sentiment(text)

impact = market_impact_score(prediction)

print("Text:", text)
print("Sentiment:", prediction["label"])
print("Confidence:", round(prediction["score"], 3))
print("Market Impact Score:", round(impact, 3))
```

```python
import json

# Notebook dosya adın burada doğru olmalı
notebook_path = "finbert_sentiment_analysis.ipynb"

with open(notebook_path, "r", encoding="utf-8") as f:
    nb = json.load(f)

# metadata.widgets varsa kaldır
if "widgets" in nb.get("metadata", {}):
    del nb["metadata"]["widgets"]

# hücrelerde widget metadata varsa temizle
for cell in nb.get("cells", []):
    if "metadata" in cell and "widgets" in cell["metadata"]:
        del cell["metadata"]["widgets"]

# Temiz notebook oluştur
clean_path = "finbert_sentiment_analysis_clean.ipynb"
with open(clean_path, "w", encoding="utf-8") as f:
    json.dump(nb, f, indent=1)

print("Temiz notebook oluşturuldu:", clean_path)
```
